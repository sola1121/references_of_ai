# 贝叶斯决策理论

贝叶斯决策理论是概率论和统计决策理论的一个分支，它提供了一种在不确定性下进行决策的数学框架。贝叶斯决策理论的核心思想是利用贝叶斯定理来更新关于某个假设的概率，即在观察到新证据后，如何调整我们对某个事件发生概率的信念。

## 关键概念

**贝叶斯定理**
贝叶斯定理是整个贝叶斯决策理论的基础，其公式如下：

    P(H|E) = [P(E|H) * P(H)] / P(E)

其中：

    P(H|E) 是后验概率，即在观察到证据 E 后，假设 H 为真的概率。
    P(E|H) 是似然概率，即在假设 H 为真的情况下观察到证据 E 的概率。
    P(H) 是先验概率，即在没有观察到任何证据之前假设 H 为真的概率。
    P(E) 是证据 E 的边缘概率，即证据 E 出现的总概率。

## 贝叶斯决策步骤

+ 定义假设空间：确定所有可能的决策选项或假设。
+ 指定先验概率：为每个假设分配一个先验概率，这通常基于历史数据或专家知识。
+ 计算似然函数：对于每个假设，计算在假设为真的情况下观察到数据的概率。
+ 应用贝叶斯定理：使用贝叶斯定理计算每个假设的后验概率。
+ 做出决策：根据后验概率选择最优的假设或决策。

## 贝叶斯决策规则

在分类问题中，贝叶斯决策规则通常遵循以下几种形式：

+ 最小化错误分类风险：选择使得期望损失最小的类别。
+ 最大后验概率（MAP）：选择具有最高后验概率的类别。
+ 最小化期望损失：选择使得期望损失最小的类别，这涉及到损失函数的定义。

## sklearn中的贝叶斯决策

在scikit-learn中，根据数据的不同分布，提供了几种不同的朴素贝叶斯分类器。

### 高斯朴素贝叶斯（Gaussian Naive Bayes）

适用于特征变量服从正态分布的数据。

    from sklearn.naive_bayes import GaussianNB
    gnb = GaussianNB()
    gnb.fit(X_train, y_train)
    y_pred = gnb.predict(X_test)

### 多项式朴素贝叶斯（Multinomial Naive Bayes）

适用于特征变量是离散的或者呈现多项式分布的数据，如文本分类问题。

    from sklearn.naive_bayes import MultinomialNB
    mnb = MultinomialNB()
    mnb.fit(X_train, y_train)
    y_pred = mnb.predict(X_test)

### 伯努利朴素贝叶斯（Bernoulli Naive Bayes）

适用于特征变量是二元分布（0或1）的数据。

    from sklearn.naive_bayes import BernoulliNB
    bnb = BernoulliNB()
    bnb.fit(X_train, y_train)
    y_pred = bnb.predict(X_test)

### 基于共轭先验的朴素贝叶斯（Complement Naive Bayes）

适用于处理不平衡数据集，是对多项式朴素贝叶斯的补充。

    from sklearn.naive_bayes import ComplementNB
    cnb = ComplementNB()
    cnb.fit(X_train, y_train)
    y_pred = cnb.predict(X_test)

在每种情况下，X_train和X_test是特征数据，y_train和y_test是相应的标签数据。使用这些分类器时，需要根据数据的特点选择合适的模型。通常，多项式朴素贝叶斯用于文本分类，高斯朴素贝叶斯用于连续数据，而伯努利朴素贝叶斯用于处理二进制/布尔数据。

# 优缺点

**优点**

+ 贝叶斯决策理论为不确定性决策提供了一个清晰的框架。
+ 它允许我们利用先验知识和数据来更新我们的信念。
+ 在数据量较少时，先验概率的引入可以减少估计的不确定性。

**缺点**

+ 计算后验概率可能非常复杂，特别是在高维数据或多个假设的情况下。
+ 先验概率的选择可能主观，且对结果有较大影响。
+ 在某些情况下，似然函数的形式可能难以确定。
+ 贝叶斯决策理论在许多领域都有应用，包括机器学习、医学诊断、经济预测等。通过贝叶斯方法，我们可以更加合理和科学地进行决策，尤其是在面对不确定性和风险时。

iris.data数据集概述：

样本数量：共150个样本。
特征属性：每个样本包含4个特征，分别是花萼长度（Sepal Length）、花萼宽度（Sepal Width）、花瓣长度（Petal Length）和花瓣宽度（Petal Width），所有特征的单位都是厘米。
类别：数据集包含3种鸢尾花，分别是Setosa（山鸢尾）、Versicolour（杂色鸢尾）和Virginica（维吉尼亚鸢尾），每种鸢尾花各有50个样本。

```python
import pandas as pd


data = pd.read_csv("./dat/iris.data", header=None)
print(data)
```

```python
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ["Fangsong"]  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号

fig = plt.figure(figsize=(10, 8))
fig.add_subplot(2, 1, 1)
plt.scatter(data[0], data[1])
plt.xlabel("花萼长度")
plt.ylabel("花萼宽度")

fig.add_subplot(2, 1, 2)
plt.scatter(data[2], data[3])
plt.xlabel("花瓣长度")
plt.ylabel("花瓣宽度")
plt.show()
```

```python
# 对数据进行切分
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(data[[0, 1, 2, 3]], data[4], test_size=0.3, random_state=0)


# 建立一个朴素贝叶斯分类器
from sklearn.naive_bayes import GaussianNB

# 拟合一个朴素叶贝斯模型
model = GaussianNB()
model.fit(x_train, y_train)

# 使用模型进行预测
y_pred = model.predict(x_test)


# 对模型进行评价
from sklearn.metrics import accuracy_score

acc = accuracy_score(y_test, y_pred)
print("精准度得分:", acc)
```
